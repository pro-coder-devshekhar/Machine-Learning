```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.datasets import make_circles
from matplotlib import pyplot
from pandas import DataFrame
import seaborn as sns
```

## <u> Question 1 </u>

```python
X,y = make_blobs(n_samples=100, random_state=1,centers=[[1, 4] ,[4, 1]], cluster_std = 0.5)

colors = ["orange", "green"]

label_name = ["Tigers", "Lions"]
fig, ax = pyplot.subplots()

df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
grouped = df.groupby('label')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=label_name[key], color=colors[key], title='dataset')
pyplot.show()
```

## <u> Question 2 </u>

```python
X,y = make_blobs(n_samples=100, random_state=1,centers=[[2, 2] ,[4, 4]], cluster_std = 0.5)

colors = ["orange", "green"]

label_name = ["ham", "spam"]
fig, ax = pyplot.subplots()

df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
grouped = df.groupby('label')

for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=label_name[key], color=colors[key], title='dataset')
ax.legend(loc='upper right')
pyplot.show()
```

## <u> Question 3 </u>

```python
X,y = make_circles(n_samples=100,shuffle=True,noise=0.05,random_state=42)

centers = [[3,4],[5,3],[4.5,6]]

X2, y2 = make_blobs(n_samples=100,cluster_std = 0.5,centers=centers,random_state=1)


for i in range(len(centers)-1,-1,-1):
    y2[y2==0+i]=i+2

print(y2)

y = np.concatenate([y,y2])

X = X*[1.2,1.8]+[3, 4]

X = np.concatenate([X,X2],axis=0)

fig,ax = pyplot.subplots()

colors = ["orange", "blue", "magenta", "yellow", "green"]

label_name = ["Tiger", "Lion", "Penguin", "Dolphin", "Python"]

for label in range(0,len(centers)+2):
    ax.scatter(X[y==label,0],X[y==label,1],c=colors[label],s=40,label=label_name[label])

ax.set(xlabel='X',ylabel='Y',title='dataset')

ax.legend(loc='upper right')

pyplot.show()
```

## <u> Question 4 </u>

```python
X, y = make_blobs(n_samples=1000, centers=4, n_features=2,random_state=100)

df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))

colors = {0:'green', 1:'orange', 2:'blue',3:'pink'}

fig, ax = pyplot.subplots()

grouped = df.groupby('label')

for key, group in grouped:
    group.plot(ax=ax,kind='scatter', x='x', y='y', label=key, color=colors[key],title='Blobs Examples')
ax.legend(loc='upper right')
pyplot.show()
```

## <u> Question 5 </u>

```python
from sklearn.datasets import fetch_olivetti_faces

faces = fetch_olivetti_faces()

faces.keys()

fig = pyplot.figure(figsize=(6, 6)) # figure size in inches

fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(faces.images[i], cmap=pyplot.cm.bone, interpolation='nearest')
    ax.text(0, 7, str(faces.target[i]))
```
